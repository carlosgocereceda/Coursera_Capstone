## 2. Scrap data from Wikipedia page into a DataFrame

```python
madridNeighborhoodList = []
data = requests.get("https://es.wikipedia.org/wiki/Anexo:Barrios_administrativos_de_Madrid").text
soup = BeautifulSoup(data, 'html.parser')
for row in soup.find_all("tr"):
    col = row.find_all('td')
    if len(col) > 0:
        if col[0].find("a"):
            madridNeighborhoodList.append(col[0].find("a").contents[0])
madridNeighborhoodList
```

```python
madrid_df = pd.DataFrame({"Neighborhood": madridNeighborhoodList})

madrid_df.head()
```

```python
coords = []
for madridNeighbor in madridNeighborhoodList:
    address = '{}, Madrid'.format(madridNeighbor)
    print(address)
    geolocator = Nominatim(user_agent="foursquare_agent")
    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude
    coords.append([latitude, longitude])
print(coords)
```

```python
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
df_coords.head()
```

<strong>Merge coords and madrid neighbor dataframes</strong>

```python
madrid_df['Latitude'] = df_coords['Latitude']
madrid_df['Longitude'] = df_coords['Longitude']
madrid_df.head()
```

```python
radius = 2000
LIMIT = 100

venues = []

for lat, long, neighborhood in zip(madrid_df['Latitude'], madrid_df['Longitude'], madrid_df['Neighborhood']):
    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))

madrid_venues_df = pd.DataFrame(venues)
madrid_venues_df.columns = ['Neighborhood', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']
madrid_venues_df.head()
```

# 5. We are going to analyze first all the venues in all the neighborhoods and then specifically for the hotel trade

```python
# one hot encoding
madrid_onehot = pd.get_dummies(madrid_venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
madrid_onehot['Neighborhoods'] = madrid_venues_df['Neighborhood']
madrid_onehot['Latitude'] = madrid_venues_df['Latitude']
madrid_onehot['Longitude'] = madrid_venues_df['Longitude']

# move neighborhood column to the first column
fixed_columns = list(madrid_onehot.columns[-3:]) + list(madrid_onehot.columns[:-3])
madrid_onehot = madrid_onehot[fixed_columns]
# madrid_onehot.head()

# now we are going to group by neighborhood
madrid_grouped = madrid_onehot.groupby(["Neighborhoods"]).mean().reset_index()

madrid_grouped.head()
```

<strong>Now let's focus only on Hotels</strong>

```python
madrid_only_hotels = madrid_grouped[["Neighborhoods","Hotel", "Latitude", "Longitude"]]
madrid_only_hotels.head()
```

# 6. Let's examine which are the neighborhoods with the most hotels

```python
madrid_grouped_total = madrid_onehot.groupby(["Neighborhoods"]).sum().reset_index()
madrid_grouped_total_hotels = madrid_grouped_total[["Neighborhoods","Hotel", "Latitude", "Longitude"]]
madrid_grouped_total_hotels.head()
```

```python
madrid_grouped_total_hotels.nlargest(5, ['Hotel']).plot.bar(x='Neighborhoods', y='Hotel', rot=0)
```

From the previous graph we can obtain interesting results such that Barajas seems to be the neighborhood with the most amount of hotels, which is easily explained since it is where the Madrid airport is located. It is likely that there are already enough hotels in this neighborhood and therefore it is not the most profitable place to establish a new one

# 7. Cluster Neighborhoods

```python
# set number of clusters
kclusters = 3

madrid_clustering = madrid_only_hotels.drop(["Neighborhoods", "Latitude", "Longitude"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(madrid_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
madrid_merged = madrid_only_hotels.copy()
madrid_merged["Cluster Labels"] = kmeans.labels_
madrid_merged.rename(columns={"Neighborhoods": "Neighborhoods"}, inplace=True)
madrid_merged.head()
```

<strong>Cluster list</strong>

```python
madrid_merged.sort_values(["Cluster Labels"])
```

# 8. Let's visualize the resulting clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(madrid_merged['Latitude'], madrid_merged['Longitude'], madrid_merged['Neighborhoods'], madrid_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-2],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
